## **Model Deployment**

### **1.1. Loading the Saved Model from Hugging Face Model Hub**

```python
# Importing required libraries

from huggingface_hub import hf_hub_download
import joblib

# Downloading model from Hugging Face Model Hub

model_path = hf_hub_download(
    repo_id="gkrishna4346/predictive-maintenance-adaboost",
    filename="adaboost_model.pkl"
)

# Loading the model

model = joblib.load(model_path)

print("Model Loaded Successfully")
print(type(model))
```

### **1.2. Getting the User Inputs**

```python
# Creating sample user inputs

engine_rpm = 1800
lub_oil_pressure = 4.5
fuel_pressure = 20.0
coolant_pressure = 3.0
lub_oil_temp = 80.0
coolant_temp = 85.0

print("Inputs Captured Successfully")
```

### **1.3. Creating Input DataFrame**

```python
# Importing pandas

import pandas as pd

# Creating input DataFrame

input_df = pd.DataFrame({
    "Engine rpm": [engine_rpm],
    "Lub oil pressure": [lub_oil_pressure],
    "Fuel pressure": [fuel_pressure],
    "Coolant pressure": [coolant_pressure],
    "lub oil temp": [lub_oil_temp],
    "Coolant temp": [coolant_temp]
})

# Displaying input DataFrame

input_df
```

### **1.4. Generating Predictions**

```python
# Generating prediction

prediction = model.predict(input_df)

# Displaying prediction

print("Predicted Engine Condition:", prediction[0])
```

### **1.5 Testing the Prediction Pipeline**

```python
# Displaying prediction result

if prediction[0] == 1:

    print("Engine Condition: Failure")
    print("Action: Maintenance Required")

else:

    print("Engine Condition: Normal")
    print("Action: No Maintenance Required")
```
