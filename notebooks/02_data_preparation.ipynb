```python
# Importing necessary libraries

import pandas as pd
from datasets import load_dataset

print("datasets library imported successfully")

# Importing train_test_split

from sklearn.model_selection import train_test_split

print("train_test_split imported successfully")
```

### **2. Load Dataset from Hugging Face Repository**

```python
# Loading dataset directly from Hugging Face

dataset = load_dataset(
    "gkrishna4346/predictive-maintenance-dataset"
)

# Display dataset structure

dataset
```

### **3. Convert Dataset to Pandas DataFrame**

```python
# Converting Hugging Face dataset into pandas DataFrame

df = dataset["train"].to_pandas()

# Display dataset dimensions

print("Dataset Shape:", df.shape)

# Display first 5 records

df.head()
```

### **4. Data Cleaning and Verification**

#### **4.1. Check Data Types**

```python
# Displaying column data types

df.dtypes
```

#### **4.2. Check Missing Values**

```python
# Checking missing values

df.isnull().sum()
```

#### **4.3. Check Duplicate Records**

```python
# Checking duplicate records

df.duplicated().sum()
```

**Observations:**

- The dataset contains 19,535 records and 7 columns.

- Data type verification confirms that all features are stored in appropriate numerical formats suitable for machine learning analysis.

- No missing values were identified across any feature.

- No duplicate records were detected in the dataset.

- All columns are relevant to the predictive maintenance problem statement and therefore no columns were removed.

- The dataset is clean, consistent, and ready for further preprocessing and model development.

### **5. Feature and Target Separation**

```python
# Separating Features and Target Variable

X = df.drop("Engine Condition", axis=1)

y = df["Engine Condition"]

# Display dimensions

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

**Observations:**

- The target variable `Engine Condition` was separated from the feature set for supervised machine learning.

- The feature matrix contains 19,535 records and 6 independent variables.

- The target variable contains 19,535 corresponding labels representing engine condition classes.

- The dataset is now ready for train-test splitting and model preparation.

### **6. Train-Test Split**

```python
# Splitting dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# Display split dimensions

print("X_train Shape:", X_train.shape)
print("X_test Shape :", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape :", y_test.shape)
```

### **7. Save Train and Test Datasets Locally**

```python
# Combining features and target for training dataset

train_df = X_train.copy()
train_df["Engine Condition"] = y_train

# Combining features and target for testing dataset

test_df = X_test.copy()
test_df["Engine Condition"] = y_test

# Saving datasets locally

train_df.to_csv("../data/train_data.csv", index=False)
test_df.to_csv("../data/test_data.csv", index=False)

print("Training and testing datasets saved successfully")
```

```python
# Verifying saved dataset dimensions

print("Train Dataset Shape:", train_df.shape)
print("Test Dataset Shape :", test_df.shape)
```

**Observations:**

- The cleaned dataset was successfully split into training and testing datasets using an 80:20 ratio.

- The feature variables and target variable were combined back together before saving.

- The training dataset contains 15,628 records, while the testing dataset contains 3,907 records.

- Both datasets were saved locally for subsequent modeling and evaluation activities.

### **8. Upload Train and Test Datasets to Hugging Face Data Space**
